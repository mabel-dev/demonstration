```python
# Import the Opteryx Library
import opteryx
```

```python
# Define SQL Statement
sql_statement: str = "SELECT * FROM $planets;"

"""
Defines the SQL query to be executed. In this example, we're selecting all
columns from the internal dataset named $planets.
"""

# Execute the SQL Query and Store the Results
results = opteryx.query(sql_statement)

"""
This function call both executes the SQL query and retrieves the results.
"""

# Display the Results
results

"""
Displays up to 10 rows of the results. This will provide you with a snapshot of
the data you queried for.

Note that there's only 9 planets, so only 9 rows are returned in this example
"""
```
